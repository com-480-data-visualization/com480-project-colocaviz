```python
year_cols = [col for col in df_prod.columns if col[0] == "Y" and col[-1] not in ["F", "N"]]
year_cols
```

```python
df_nans = df_prod.isna().sum()
```

```python
df_nans = df_nans[year_cols].rename(lambda x: x[1:])
```

```python
sns.lineplot(df_nans)
plt.title("Number of nans by year")
plt.xlabel("Year")
plt.ylabel("Number of Nans")
plt.xticks(ticks=df_nans.index[::5], rotation=45)
plt.show()
```

```python
df_prod.groupby(by="Area")[year_cols].apply(lambda x: x.isna().sum()).sum(axis=1) / df_prod.groupby(by="Area").size()
```

```python
df_prod.groupby(by="Area").size()
```

```python
nans_by_country = (df_prod.groupby(by="Area")[year_cols].apply(lambda x: x.isna().sum()).sum(axis=1) / df_prod.groupby(by="Area").size()).sort_values(axis=0, ascending=False)

top_nb = 5
indexes = [i for i in range (top_nb)] + [-top_nb+i for i in range(top_nb)]

top = pd.DataFrame(nans_by_country.iloc[indexes]).rename(columns={0: 'nb_nans'})
display(top)
top["is_top"] = [True]*top_nb + [False]*top_nb
top
```

```python
nans_by_country
```

```python
sns.barplot(data=top, x="Area", y="nb_nans", hue="is_top")
plt.title(f"Top and bottom {top_nb} areas by number of missing values per row")
plt.xticks(rotation=90)
plt.xlabel("Area")
plt.ylabel("Number of Nans per row")
plt.show()
```

```python
sns.lineplot(data=pd.DataFrame(nans_by_country), x="Area", y=0)
plt.title("Number of nans by area")
plt.xticks(ticks=pd.DataFrame(nans_by_country).index[::10], rotation=90)
plt.xlabel("")
plt.ylabel("Number of Nans by row")
plt.show()
```

```python
df_prod.groupby(by=['Item'])[year_cols].apply(lambda x: x.isna().sum()).sum(axis=1).sort_values(axis=0, ascending=False)
```

```python
df_prod.groupby(["Element"]).size()/df_prod.groupby(["Element", "Unit"]).size().sum()
```

```python
prod_df = df_prod[(df_prod['Element'] == 'Production') & (df_prod['Unit'] == 't')]
prod = prod_df.groupby(by="Item")[year_cols].sum().sum(axis=1)
```

```python
area_df = df_prod[(df_prod['Element'] == "Area harvested")]
area = area_df.groupby(by="Item")[year_cols].sum().sum(axis=1)
```

```python
prod_area = pd.concat([prod, area], axis = 1).dropna().rename(columns={0: "Production", 1: "Area"})
sns.scatterplot(prod_area, x="Production", y="Area")
plt.title("Total Production against Area used for different items")
plt.xlabel("Production in t")
plt.ylabel("Area used in ha")
```
